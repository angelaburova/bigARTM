```python
T = 12
subjects_names = ["sbj"+str(i) for i in range(T)]
print subjects_names
table = []
for sub in subjects_names:
    tmp = phi.get(sub)
    x = []
    for i in range(tmp):
        x.append(x[i])
    table.append(x)

num_of_subs = len(table)
num_of_words = len(table[0])
print num_of_subs
print num_of_words

for i in range(num_of_words):
    avrg = 0.0
    for k in range(num_of_subs):
        avrg += table[k][i]
    for k in range(num_of_subs):
        table[k][i] -= avrg
        if table[k][i]<0:
            table[k][i] = 0.0
```
